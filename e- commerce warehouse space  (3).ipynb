```python
dcols=[  'Total_Space_Used', 'Utilization_Rate', 'Total_Cost', 'Total_Revenue', 'Profit']
```

```python
dcols
```

```python
sns.boxplot(dspace["Quantity_Stored"])
```

```python
for i in dcols:
    skew=dspace[i].skew().round()
    print(skew)
```

```python
for i in dcols:
    q1=dspace[i].quantile(0.25)
    q3=dspace[i].quantile(0.75)
    iqr=q3-q1
    upperlimit=q3+1.5*iqr
    lowerlimit=q1-1.5*iqr
    dspace[i]=dspace[i].map(lambda x: upperlimit if x>upperlimit else lowerlimit if x<lowerlimit else x)
```
